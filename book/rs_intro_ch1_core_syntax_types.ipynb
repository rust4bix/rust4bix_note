# Core Syntax & Types

## Why Syntax Matters in Bioinformatics

In bioinformatics, we routinely process gigabyte–terabyte scale data (FASTQ, BAM, VCF). Small inefficiencies in syntax or type usage can translate into hours of runtime or gigabytes of wasted memory.

Rust’s syntax is not just aesthetic—it enforces:

* Memory safety (no silent bugs)

* Explicit data handling

* High performance without a garbage collector

This chapter builds the foundation for writing correct and fast genomic pipelines.


## Variables and Immutability


Rust variables are immutable by default.


````rust
let x = 10;
// x = 20; // error
````

If you run the Rust code below: 

```{rust}
let x = 10;
x = 20; // error
```

You will get the error message shown below: 



## Mutable variables


````rust
let mut count = 0;
count += 1;
println!("count: {}", count);
````

Use `mut` only when necessary—this reduces bugs and improves reasoning.

## Shadowing (Rebinding Variables)



````rust
let x = 5;
let x = x + 1;
let x = x * 2;
println!("x: {}", x);
````

Unlike mutation, shadowing creates a new variable.

````rust
let seq = "acgt";
let seq = seq.to_uppercase();
println!("seq: {}", seq);
````

You transform data without mutating the original reference.

### Integers

````rust
let x: i32 = 10;
let y: u64 = 100;
````

Use cases:

* `u64` → genome positions

* `i32` → score differences

````rust
let gc_ratio: f64 = 0.42;
````
