# Generics functions and structs



````rust
use std::fmt::Display;
use std::fmt::Debug;

fn highest_expr<T: PartialOrd + Copy + Display + Debug> (list: &[T]) -> T {
    let mut highest = list[0];
    for &item in list.iter() {
        println!("item: {}", item);
        println!("debug format: {:?}", item);   // debug format (also 3 here)
        println!("display: {}", &item);
        println!("pointer: {:p}", &item); // prints memory address
        if item > highest {
            highest = item;
        }
    }
    highest
}

let numbers = [3, 1, 4, 1, 5, 9];
let highest_number = highest_expr(&numbers);
println!("The highest number is: {}", highest_number);

let expr_in_float = [3.0, 1.0, 4.0, 1.0, 5.0, 9.0];
let highest_number = highest_expr(&expr_in_float);
println!("The highest number is: {}", highest_number);
````

We are doing pattern matching (destructuring):

* `list.iter()` gives `&T`

* `&item` pattern unwraps that reference

* so inside the loop: `item: T`
