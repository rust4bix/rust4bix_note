## Transforming Data with `map()`


The `map` function is one of the most commonly used transformation tools in Rust. It applies a function to each value in a container or wrapper type and returns a new value with the transformation applied.


In Rust, map appears in several places, most commonly:

* **Iterators** (   Iterator::map)
* **Option** (Option::map)
* **Result** (Result::map)

```rust
use std::collections::HashMap;

/// A closed genomic interval: both `start` and `end` are included.
/// In bioinformatics this is common, so `100..100` means one base.
#[derive(Clone, Debug, PartialEq, Eq)]
pub struct Interval {
    pub start: u64,
    pub end: u64,
}

impl Interval {
    pub fn new(start: u64, end: u64) -> Self {
        Self { start, end }
    }

    pub fn len(&self) -> u64 {
        // Genomic intervals here are inclusive on both ends.
        // Example: 10..12 contains 3 bases: 10, 11, and 12.
        self.end - self.start + 1
    }
}
```

```rust
let mut intervals: Vec<Interval> = vec![
    Interval::new(10, 12),
    Interval::new(20, 25),
];
println!("{:?}", intervals);
println!("Length of first interval: {}", intervals[0].len());
println!("Length of second interval: {}", intervals[1].len());
```

```rust
// append new intervals
intervals.push(Interval::new(20, 35));
println!("{:?}", intervals);
println!("The number of intervals: {}", intervals.len());
```
