## Transforming Data with `map()`


The `map` function is one of the most commonly used transformation tools in Rust. It applies a function to each value in a container or wrapper type and returns a new value with the transformation applied.


In Rust, map appears in several places, most commonly:

* **Iterators** (   Iterator::map)
* **Option** (Option::map)
* **Result** (Result::map)

```rust
let numbers = vec![1, 2, 3, 4];

let doubled: Vec<i32> = numbers
    .iter()
    .map(|x| x * 2)
    .collect();
println!("Original: {:?}", numbers);
println!("After doubling: {:?}", doubled);
```
